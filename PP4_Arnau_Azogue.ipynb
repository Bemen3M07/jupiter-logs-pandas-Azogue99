# Exercici 2

### Per que creus que es necessari implementar un log en un programa? Raona la resposta i fica exemples del perquè ho creus (mínim 3 exemples).
---
Implementar un sistema de **logs** en un programa és necessari perquè ajuda a fer un seguiment detallat del comportament i l'estat del programa durant l'execució. Els **logs** proporcionen informació útil per depurar errors, supervisar el rendiment i entendre el flux del codi, especialment en aplicacions grans i complexes. Aquí tens tres raons amb exemples per il·lustrar-ho:  
1. **Depuració d'Errors**: Quan es produeix un error inesperat, un sistema de logs pot ajudar a identificar el punt exacte del problema, registrant el missatge d'error i el context en què va ocórrer.
   - *Exemple*: Si un programa falla en llegir un fitxer de configuració, un log pot indicar l'error específic ("Error: No es pot trobar el fitxer config.json") i la línia de codi on es va produir.
<br><br>
2. **Monitorització del Rendiment**: Els logs es poden utilitzar per mesurar el temps d'execució de funcions o operacions crítiques, ajudant a identificar colls d'ampolla o rendiment baix.
   - *Exemple*: Si una aplicació web té temps de càrrega lents, es poden afegir logs per registrar el temps de resposta de les peticions, detectant quines parts del sistema triguen més.
<br><br>
3. **Seguretat i Auditoria**: Els logs ajuden a rastrejar i registrar accions sensibles o sospitoses en un programa, com accessos no autoritzats, intents de connexió, o modificacions crítiques de dades.
   - *Exemple*: En una aplicació bancària, els logs poden registrar cada intent d'inici de sessió, els canvis en el saldo d'un compte o qualsevol accés als registres financers per auditar possibles anomalies o fraus.
<br><br>
---
Així doncs, els logs proporcionen un registre detallat que és fonamental per millorar la fiabilitat, el rendiment i la seguretat d'un programa.

### Amb les teves paraules, explica que significa cada línia dels següents exemples de logging. CAL EXPLICAR LINEA A LINEA

### Logging 1

```python
logging.basicConfig(format='%(asctime)s - %(message)s', level=logging.INFO)
```
- Aquesta línia configura el sistema de logging amb un format específic i un nivell de registre.
  - `format='%(asctime)s - %(message)s'`: Estableix el format dels missatges de log. Inclou la data i l'hora (`%(asctime)s`) seguit del missatge (`%(message)s`).
  - `level=logging.INFO`: Indica que només es registraran els missatges de nivell `INFO` i superior (com `WARNING`, `ERROR`, etc.).

```python
logging.info('Admin ha accedido')
```
- Registra un missatge d'informació (`INFO`). El text `'Admin ha accedido'` s'enregistrarà segons el format especificat, incloent la marca de temps.

---

### Logging 2

```python
f_handler = logging.FileHandler('file.log')
```
- Aquesta línia crea un **handler** que escriurà els logs en un fitxer anomenat `'file.log'`.
  - `logging.FileHandler('file.log')`: Crea un objecte que maneja l'escriptura dels missatges de log en un fitxer.

---

### Logging 3

```python
import logging
```
- Aquesta línia importa el mòdul de logging de Python per habilitar la funcionalitat de registre de missatges.

```python
logging.basicConfig(format='%(process)d-%(levelname)s-%(message)s')
```
- Configura el format dels missatges de log.
  - `%(process)d`: Inclou l'ID del procés que està executant el codi.
  - `%(levelname)s`: Afegeix el nivell del missatge (ex: `WARNING`, `INFO`).
  - `%(message)s`: El missatge que es vol registrar.

```python
logging.warning('This is a Warning')
```
- Registra un missatge de nivell `WARNING` amb el text `'This is a Warning'`. Aquest missatge es mostrarà segons el format especificat i només es registrarà perquè el nivell `WARNING` està per sobre de `INFO`.
